```python
import cv2
```

```python
import cvzone
```

```python
from cvzone.SelfiSegmentationModule import SelfiSegmentation
```

```python
import os
```

```python
cap=cv2.VideoCapture(0)
```

```python
cap.set(3,640)
cap.set(4,480)
cap.set(cv2.CAP_PROP_FPS,60)
segmentor = SelfiSegmentation()
fpsReader= cvzone.FPS()
# imgbg=cv2.imread("C:/Users/KIIT/Documents/Projects/Background Remover/images/f.jpg")

listimg = os.listdir("C:/Users/KIIT/Documents/Projects/Background Remover/images")
imglist = []

for imgpath in listimg:
    img=cv2.imread(f'C:/Users/KIIT/Documents/Projects/Background Remover/images/{imgpath}')
    imglist.append(img)
indeximg = 0
while True:
    success, img = cap.read()
    imgOut = segmentor.removeBG(img, imglist[indeximg], threshold=0.75)
    imgst = cvzone.stackImages([img,imgOut],2,1)
    _ ,fpsReader.update(imgst,color=(0,0,255))
    print(indeximg)
    cv2.imshow("Stacked Img",imgst)
    key=cv2.waitKey(1)
    if key == ord('a'):
        if indeximg>0:
            indeximg-=1
    elif key == ord('d'):
        if indeximg <len(imglist)-1:
            indeximg+=1
    elif key == ord('q'):
        break
```
